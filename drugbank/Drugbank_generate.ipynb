```python
with open("protein_drugbank.json") as file:
    data = file.read()
    data = json.loads(data)

uniprot_ids = []
for k, v in data.items():
    if (v): uniprot_ids.append(k)
```

```python
len(uniprot_ids)
```

```python
protein_drugs = {}
i = 1
for uniprot_id in uniprot_ids:
    try:
        drugs = get_drugbank_info(uniprot_id)
        protein_drugs[uniprot_id] = drugs
        print(f"{i}: Got information about {uniprot_id} in drugbank")
        i += 1
    except:
        protein_drugs[uniprot_id] = None
        print(f"No information about {uniprot_id} in drugbank")

# 408-409 issue!
```

```python
json_data = json.dumps(protein_drugs)
with open("protein_drugbank.json", "w") as file:
    file.write(json_data)
file.close()
```
