```python
import json 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

with open("../protein_properties/proteins.json") as f:
    data = f.read()
    proteins = json.loads(data)
len(proteins)
```

```python
mec_280 = [info["Molar Extinction Coefficient"][0] for info in proteins.values()]
mec_205 = [info["Molar Extinction Coefficient"][1] for info in proteins.values()]
gravy = [info["GRAVY"] for info in proteins.values()]
instability = [info["Instability Index"] for info in proteins.values()]
isoelectric = [info["Isoelectric Point"] for info in proteins.values()]
molecular_weight = [info["Molecular Weight"] for info in proteins.values()]
aromaticity = [info["Aromaticity"] for info in proteins.values()]
```

```python
# Create a dataframe with all the properties
df = pd.DataFrame({
    "MEC 280 nm": mec_280,
    "MEC 205 nm": mec_205,
    "GRAVY": gravy,
    "Instability Index": instability,
    "Isoelectric Point": isoelectric,
    "Molecular Weight": molecular_weight,
    "Aromaticity": aromaticity
})
df.head()
```

```python
import seaborn as sns

# Compute the correlation matrix
corr = df.corr()
# Heatmap
sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns,
        cmap="Blues")
```

```python
# Pairplot
sns.pairplot(df)
```
